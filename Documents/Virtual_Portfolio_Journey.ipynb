# Creating Virtual Environment

Why are we using a virtual environment?

To host the project in one place and manage the version of code used as well as versions of packages imported. 

- Open Mac terminal
- cd to location you want the virtual environment to be saved
- Create virtual environment using the following command:

conda create --name 'vitural_portfolio_env' python3.10 anaconda

- Activate the virtua environment using command:

conda activate 'virtual_portfolio_env'

**Official doc:** https://docs.conda.io/projects/conda/en/latest/user-guide/tasks/manage-environments.html#:~:text=To%20set%20environment%20variables%2C%20run,conda%20env%20config%20vars%20list%20.


# Useful explanation of the Git/Anaconda workflow

https://tdhopper.com/blog/my-python-environment-workflow-with-conda

Steps are

1) Create project folder

2) Create Anaconda virtual env - "conda create --name 'virtual_portfolio_env' psycopg2 numpy plotly dash pandas matplotlib typing"

3) activate conda env - "conda activate virtual_portfolio_env"

4) Create config.yaml file for env - "conda env export > virtual_portfolio_environment.yaml
"

5) find path where env stored 'conda env list'. Then cd to that path

6) creating 2 separate directories to store an environment variables file. one for activating environment variables, one for deactivating.

5) (optional) create an env file containing a line to auto activate the env whenever the project folder is navigated to from the terminal

6) run $ git init to make the folder a Git repository. "I then run $ git add environment.yml && git commit -m 'initial commit' to add the YAML file to the repository."

7) pushing the repository to github, then push the master branch: $ git create, then $ git push -u origin master.
